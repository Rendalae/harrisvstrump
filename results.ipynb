```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/Users/aaronviviani/code/aaronviviani/08 - Project/1976-2020-president.csv")
df
```

```python
state_groups = df.groupby(['year', 'state'])

def determine_majority(group):

    max_votes = group['candidatevotes'].max()

    group['majority'] = group['candidatevotes'] == max_votes
    return group

df = state_groups.apply(determine_majority)

df.reset_index(drop=True, inplace=True)
```
